```python
# Methods = functions
# An object type=Tree
class Tree(object):
    
    worker1='Tom'
    worker2='Jimmy'
    
    def __init__(self,tree_name):
        self.tree_name = tree_name
    
    def tree_water(self):
        print(("The {} tree is being Watered by {}".format(self.tree_name,Tree.worker1)))
    
    def tree_harvest(self):
        print(("The {} tree is being Harvested by {}".format(self.tree_name,Tree.worker2)))

# creation of class instance tr   
# just like list_1 was an instance of object list

tr = Tree(tree_name='palm')
print(type(tr))
print(tr.tree_name)
tr.tree_harvest()



tr = Tree(tree_name='apple')  #tree_name is an attribute
print(type(tr))
print(tr.tree_name)
tr.tree_water()  #tree_water is a method
```

```python
def reverse_string(string):
  """Reverses a string.

  Args:
    string: The string to reverse.

  Returns:
    The reversed string.
  """
  reversed_string = string[::-1]
  return reversed_string
```

```python
reversed_string = reverse_string("Title")
print(reversed_string)
```

```python
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b

def divide(a, b):
  return a / b

while True:
  print("Select operation.")
  print("1. Add")
  print("2. Subtract")
  print("3. Multiply")
  print("4. Divide")

  operation = input("Select operation(1/2/3/4): ")

  if operation in ('1', '2', '3', '4'):
    first_number = float(input("Enter first number: "))
    second_number = float(input("Enter second number: "))

    if operation == '1':
      print(first_number, "+", second_number, "=", add(first_number, second_number))
    elif operation == '2':
      print(first_number, "-", second_number, "=", subtract(first_number, second_number))
    elif operation == '3':
      print(first_number, "*", second_number, "=", multiply(first_number, second_number))
    elif operation == '4':
      print(first_number, "/", second_number, "=", divide(first_number, second_number))

  else:
    print("Invalid input!")

  choice = input("Continue calculation? (Y/N): ")
  if choice == "N":
    break
```

```python
def find_longest_common_substring(string1, string2):
  """Finds the longest common substring between two strings.

  Args:
    string1: The first string.
    string2: The second string.

  Returns:
    The longest common substring between the two strings.
  """
  # Create a table to store the lengths of longest common suffixes of substrings.
  lcs_length_table = [[0 for _ in range(len(string2) + 1)] for _ in range(len(string1) + 1)]

  # Initialize the table.
  for i in range(len(string1) + 1):
    for j in range(len(string2) + 1):
      if i == 0 or j == 0:
        lcs_length_table[i][j] = 0
      elif string1[i - 1] == string2[j - 1]:
        lcs_length_table[i][j] = lcs_length_table[i - 1][j - 1] + 1
      else:
        lcs_length_table[i][j] = 0

  # Find the maximum length of a common suffix.
  max_lcs_length = 0
  for i in range(len(string1) + 1):
    for j in range(len(string2) + 1):
      if lcs_length_table[i][j] > max_lcs_length:
        max_lcs_length = lcs_length_table[i][j]

  # Find the start and end indices of the longest common substring.
  start_index = 0
  end_index = 0
  for i in range(len(string1) + 1):
    for j in range(len(string2) + 1):
      if lcs_length_table[i][j] == max_lcs_length:
        start_index = i - lcs_length_table[i][j]
        end_index = i
        break

  # Return the longest common substring.
  return string1[start_index:end_index]
```

```python
string1 = "hello world"
string2 = "world"

longest_common_substring = find_longest_common_substring(string1, string2)

print(longest_common_substring)
```
