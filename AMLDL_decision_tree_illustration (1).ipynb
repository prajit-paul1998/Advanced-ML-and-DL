```python
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
from sklearn import datasets, tree
```

```python
df = pd.read_csv('example_dist.csv')
```

```python
df
```

```python
df['Physics'] = df['Physics'].replace({'N': 0, 'Y': 1})
df['Chemistry'] = df['Chemistry'].replace({'N': 0, 'Y': 1})
df['Maths'] = df['Maths'].replace({'N': 0, 'Y': 1})
df['Result'] = df['Result'].replace({'N': 0, 'Y': 1})
```

```python
df
```

```python
# Split the data into features and target
X = df[['Physics', 'Chemistry', 'Maths']]
y = df['Result']
```

```python
X
```

```python
y
```

```python
# Create a decision tree classifier
clf = tree.DecisionTreeClassifier()
```

```python
# Fit the classifier to the data
clf = clf.fit(X, y)
```

```python
# Visualize the decision tree
plt.figure()
tree.plot_tree(clf)
plt.show()
```
