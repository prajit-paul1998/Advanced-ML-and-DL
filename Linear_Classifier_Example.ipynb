```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_excel('my_own_dataset.xlsx')
```

```python
df1
```

```python
X = df1[['x1','x2','x3','x4']]
print(X)
```

```python
y = df1[['label']]
print(y)
```

```python
x_var=X.to_numpy() 
print(x_var)
```

```python
y_var=y.to_numpy()
print(y_var)
```

```python
y_var.ravel()
```

```python
# Create logistic regression Model
model1 = linear_model.LogisticRegression(solver='liblinear',C=100)
```

```python
# Train the classifier
model1.fit(x_var,y_var.ravel())
```

```python
# Testing of model1
pred_class = model1.predict(np.array([[1,3,3,4]]))
print(pred_class)
```
